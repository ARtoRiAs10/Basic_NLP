```python
import nltk
nltk.download('punkt')
```

```python
corpus = """Ticket is mandatory for 3 years+ aged children.
 Person below 18 years cannot be admitted for `A` certified movies.
 Items like laptops, cameras, knives, lighter, matchboxes, cigarettes, firearms, and inflammable objects are strictly prohibited.
 Items like carrying bags, eatables, helmets, and handbags are strictly prohibited inside the theatre and should be deposited at the baggage counter of the mall /cinema.
 3D movie ticket price includes 3D glass usage charges.
 Seat layout page of Audi is for representational purposes and the actual seat layout might vary.
 Outside food & beverages are not allowed on the cinema premises.
 Decision(s) taken by Cinepolis management is final & abiding.
 Purchased tickets cannot be cancelled, exchanged, or refunded.
 Cinepolis may contact the guest to seek feedback for service improvement.
 Patrons under the influence of alcohol or drugs will not be allowed inside the cinema premises.
 Pre-booked food & beverage needs to be collected from F&B counter (except in VIP Cinemas)
 Wearing the mask is mandatory (as per guidelines of the local authority)
Movie schedules and rates are subject to change without prior notice."""
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
#Paragraph -> word
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
corpus2=(tokenizer.tokenize(corpus))
```

```python
print(corpus2)
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
detokenizer=TreebankWordDetokenizer()
```

```python
detokenizer.tokenize(corpus2)
```
